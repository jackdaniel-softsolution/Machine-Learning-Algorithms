# one D scatter plot

```python
setosa=df.loc[df["species"]=="setosa"] 
versicolor=df.loc[df["species"]=="versicolor"] 
virginica=df.loc[df["species"]=="virginica"] 

plt.plot(setosa["petal_length"],np.zeros_like(setosa["petal_length"]))
plt.plot(versicolor["petal_length"],np.zeros_like(versicolor["petal_length"]))
plt.plot(virginica["petal_length"],np.zeros_like(virginica["petal_length"]))

plt.show()
```

# CDF

```python
counts,bin_edges=np.histogram(setosa["petal_length"],bins=10,density=True)
```

```python
print(counts)
print(bin_edges)
```

```python
pdf=counts/sum(counts)
```

```python
print(pdf)
print(bin_edges)
```

```python
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
#setosa

counts,bin_edges=np.histogram(setosa["petal_length"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
#versicolr
counts,bin_edges=np.histogram(versicolor["petal_length"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
#verginica
counts,bin_edges=np.histogram(virginica["petal_length"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

# Mean,Std

```python
print(np.mean(setosa["petal_length"]))
print(np.mean(versicolor["petal_length"]))
print(np.mean(virginica["petal_length"]))
```

```python
#the mean tells the all setosa flower average petal_length is 1.46
print(np.std(setosa["petal_length"]))
print(np.std(versicolor["petal_length"]))
print(np.std(virginica["petal_length"]))
```

# Median

```python
print(np.median(setosa["petal_length"]))
print(np.median(versicolor["petal_length"]))
print(np.median(virginica["petal_length"]))
```
