```python
#style a graph


from matplotlib import style 
style.use("ggplot")
slices=[10,40,50,80]
lable=['high salary','medium salary','low salary','umemployement']
plt.pie(slices,labels=lable)
plt.show
```

```python
#style a graph


from matplotlib import style 
style.use("ggplot")
slices=[10,40,50,80]
lable=['high salary','medium salary','low salary','umemployement']
col=['blue','yellow','orange','black']
plt.pie(slices,labels=lable,shadow=True,explode=(0,0,0,0.1),autopct='%1.1f%%') #explode--> highlight the slice
plt.show                                                                       #autopact--->give the percentage 
```

```python
#style a graph


from matplotlib import style 
style.use("ggplot")
x1=[1,2,3,4]
y1=[5,6,7,8]
x2=[9,10,11,12]
y2=[13,14,15,16]
plt.subplot(211)
plt.plot(x1,y1)
plt.title("multiple plot")
plt.ylabel("first graph")

plt.subplot(212)
plt.plot(x2,y2)
plt.ylabel("second graph")
```
