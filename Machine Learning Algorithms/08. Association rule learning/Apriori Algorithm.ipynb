```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Market_Basket_Optimisation.csv", header = None)
```

```python
data
```

```python
new_data=[]
for i in range(7501):
    new_data.append([str(data.values[i,j]) for j in range(20)])
```

```python
new_data
```

```python
from apyori import apriori
rules=apriori(new_data,min_support=0.003,min_confidence=0.2,min_lift=2,min_length=2)
```

```python
results=list(rules)
```

```python
results
```

```python
lift=[]
association=[]
for i in range(len(results)):
    lift.append(results[:len(results)][i][2][0][3])
    association.append(list(results[:len(results)][i][0]))
```

```python
conclusion=pd.DataFrame([association,lift]).T
conclusion.columns=["association","lift"]
```

```python
conclusion.sort_values("lift", ascending=False)
```
